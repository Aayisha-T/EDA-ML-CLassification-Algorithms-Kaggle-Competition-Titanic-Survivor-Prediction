```python
Age_range=[] 
survived=[]
not_survived= []
total=[]
for x in range(0,76,5):
    s=len(train[(train.Age>x)&(train.Age<=x+5)&(train.Survived==1)])
    n=len(train[(train.Age>x)&(train.Age<=x+5)&(train.Survived==0)])
    tot=len(train[(train.Age>x)&(train.Age<=x+5)])
    y=x+5
    a=f'{x} - {y}'
    survived.append(s)
    not_survived.append(n)
    Age_range.append(a) 
    total.append(tot)
age_survival=pd.DataFrame({'Age Range':Age_range, 'Survived':np.array(survived)/np.array(total)*100, 'Not Survived': np.array(not_survived)/np.array(total)*100})
```

```python
np.array(survived)/np.array(total)*100,2
```

```python
age_survival.sort_values(by=['Survived'], ascending=False)
#0-15 and 75+ more survived then not
```
